<h4>Exploratory Data analysis

```python
df["Order Date"].min()
```

```python
df["Order Date"].max()
```

```python
df["month_year"] = df["Order Date"].apply(lambda x: x.strftime("%Y-%m"))
```

```python
df_trend = df.groupby("month_year").sum()["Sales"].reset_index()
df_trend.head()
```

```python
plt.figure(figsize = (15,6))
plt.plot(df_trend["month_year"], df_trend["Sales"], color = "green")

plt.xticks(rotation = "vertical", size = 8)
plt.xlabel("Month_Year")
plt.ylabel("Sales")
plt.show();
```

<h4> Which are the top 10 products by sales

```python
#Grouping product name by column
prod_sales = pd.DataFrame(df.groupby("Product Name").sum()["Sales"])
```

```python
prod_sales = prod_sales.sort_values("Sales", ascending = False)
```

```python
prod_sales[:10]
```

<h4>Most selling products

```python
#Group product name
prod_most_sold = df.groupby("Product Name").sum()["Quantity"]
```

```python
prod_most_sold = prod_most_sold.sort_values(["Quantity"], ascending = False)
```

```python
prod_most_sold[:10]
```

<h4>Which are the most profitable category and sub-category

```python
cat_subcat_profit = pd.DataFrame(df.groupby(["Category", "Sub-Category"]).sum()["Profit"])
```

```python
cat_subcat_profit.sort_values(["Category", "Profit"], ascending = False)
```
